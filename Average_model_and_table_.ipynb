```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df1=pd.read_csv("/content/test_output_1.csv")
df2=pd.read_csv("/content/test_output_exp.csv")
```

```python
obs=df1["observed"].tolist()
lin_pred=df1["lin_predicted"].tolist()
exp_pred=df2["exp_predicted"].tolist()
```

Averaging the prediction of linear regression and exponential smoothing models

```python
avg_pred=[]
for i in range (len(obs)):
  avg_pred.append(((lin_pred[i])+(exp_pred[i]))/2)
print(avg_pred)
```

Observing different kind of error associated with average models

```python
#perfect prediction and recent prediction

plt.scatter(obs, avg_pred)
plt.plot(obs,obs, color='r')

model = LinearRegression()
obs = np.array(obs).reshape(-1, 1)
model.fit(obs, np.array(avg_pred))
y_pred = model.predict(obs)
plt.plot(obs,y_pred, color='g')


plt.xlabel("obs")
plt.ylabel("lin_pred")
plt.show()
```
